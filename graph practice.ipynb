```python
#contour map
import matplotlib 
import numpy as np
import matplotlib.cm as cm #color maps
import matplotlib.pyplot as plt 

delta =0.025
x = np.arange(-3.0,0.3, delta)
y = np.arange(-2.0,0.2, delta)
X, Y =np.meshgrid(x, y)
Z1=np.exp(-X**2 - Y**2)
Z2=np.exp(-(X - 1)**2 - (Y - 1)**2)
Z=(Z1 - Z2) * 2

print(X)
print(Y)
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X, Y ,Z)
ax.clabel(CS, inline=1, fontsize=10)
ax.set_title('contour map')
```
